```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import warnings
# !pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
data.groupby('Outcome').mean()
```

```python
X = data.drop(columns=['Outcome'], axis=1)
y = data['Outcome']
```

```python
# print(X)
# print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, stratify=y, random_state=42)
print(X.shape, X_train.shape, X_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

```python
model = LogisticRegression()
modelR = RandomForestClassifier()
modelS = SVC()
modelK = KNeighborsClassifier()
```

```python
model.fit(X_train,y_train)
modelR.fit(X_train,y_train)
modelK.fit(X_train,y_train)
modelS.fit(X_train,y_train)
```

```python
X_train_prediction = model.predict(X_train)
X_train_predictionR = modelR.predict(X_train)
X_train_predictionK = modelK.predict(X_train)
X_train_predictionS = modelS.predict(X_train)
```

```python
X_test_predictionL = model.predict(X_test)
X_test_predictionK = modelK.predict(X_test)
X_test_predictionR = modelR.predict(X_test)
X_test_predictionS = modelS.predict(X_test)
```

```python
acc_trainL = accuracy_score(X_train_prediction,y_train)
acc_trainR = accuracy_score(X_train_predictionR,y_train)
acc_trainK = accuracy_score(X_train_predictionK,y_train)
acc_trainS = accuracy_score(X_train_predictionS,y_train)
acc_testL = accuracy_score(X_test_predictionL, y_test)
acc_testR = accuracy_score(X_test_predictionR, y_test)
acc_testK = accuracy_score(X_test_predictionK, y_test)
acc_testS = accuracy_score(X_test_predictionS, y_test)



print(f"Accuracy on Training data by Logistic Regression: {acc_trainL}")
print(f"Accuracy on Testing data by Logistic Regression: {acc_testL}")

print(f"\nAccuracy on Training data by Random Forest: {acc_trainR}")
print(f"Accuracy on Testing data by Random Forest: {acc_testR}")

print(f"\nAccuracy on Training data by KNN: {acc_trainK}")
print(f"Accuracy on Testing data by KNN: {acc_testK}")

print(f"\nAccuracy on Training data by SGD Classifier: {acc_trainS}")
print(f"Accuracy on Testing data by SGD Classifier: {acc_testS}")
```
